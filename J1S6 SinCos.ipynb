# Create a Jupyter Notebook where, in separate cells, you define functions that return sin(x) and cos(x). Use Markdown cells to comment your Notebook, and describe what each function does. Create a third Python cell that will tabulate sin(x) and cos(x) using these previously defined functions vs. x, where x is tabulated between 0 and 2pi with a thousand entries. Write a fourth Python cell that will use a for loop to print out the first 10 values of x, sin(x), and cos(x) in columns.

```python
import math
```

calcCos takes the parameter x and returns its cosine value.

```python
def calcCos(x):
    return math.cos(x)
```

calcSin takes the parameter x and returns its sine value.

```python
def calcSin(x):
    return math.sin(x)
```

makeTable makes the table for sine and cosine values of x. Doesn't print it yet.

```python
def makeTable(startPt, endPt, entries):
    table = []
    step = (endPt - startPt) / (entries - 1)
    #    =  6.28 / 999
    for i in range(entries):
        x = startPt + i * step
        sinOfX = calcSin(x)
        cosOfX = calcCos(x)
        table.append((x, sinOfX, cosOfX))
    return table
```

printTable prints the table with specified number of rows.

```python
def printTable(table, rows):
    for x, sinOfX, cosOfX in table:
        #print(rows)
        print(f"x = {x}, sin(x) = {sinOfX}, cos(x) = {cosOfX}")
        rows -= 1
        if rows <= 0:
            break
```

```python
def main():
    startPt = 0
    endPt = 2*(math.pi)
    entries = 1000

    table = makeTable(startPt, endPt, entries)
    printTable(table, 10)

if __name__ == "__main__":
    main()
```
