```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def exponent(a, b, x):
    return a * np.exp(-b * x)
```

```python
def makeArray(startPt, endPt, numPts):
    step = (endPt - startPt) / (numPts - 1)
    expArray = []
    for i in range(numPts):
        x = startPt + i * step
        y = exponent(4, 5, x) # Chosen values for a and b
        expArray.append((x, y))
    return expArray
```

```python
def main():
    startPt = 0
    endPt = 2 * np.pi
    numPts = 100
    
    array = makeArray(startPt, endPt, numPts)
    #xVals = [x for x, y in array]
    yVals = [y for x, y in array]
    
    plt.plot(yVals)
    plt.xlabel("X Axis")
    plt.ylabel("Y Axis")
    plt.title("Kevin Bozarth Journal 2 Session 1")
    plt.savefig("J2S1 Graph.pdf")
    plt.show()
```

```python
if __name__ == "__main__":
    main()
```
