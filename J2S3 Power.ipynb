```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def exponent(a, b, x):
    return a * np.exp(-b * x)
```

```python
def makeArray(startPt, endPt, numPts):
    step = (endPt - startPt) / (numPts - 1)
    sigmaStep = (0.2 - 2) / (numPts - 1)
    expArray = []
    for i in range(numPts):
        x = startPt + i * step
        sigma = 2 + i * sigmaStep
        y = exponent(4, 5, x)
        scatter = np.random.normal(0, sigma)
        yScatter = y + scatter
        expArray.append((x, yScatter, sigma))
    return expArray
```

```python
def main():
    startPt = 0
    endPt = 2 * np.pi
    numPts = 100
    
    array = makeArray(startPt, endPt, numPts)
    xVals = [x for x, y, sigma in array]
    yVals = [y for x, y, sigma in array]
    sigmas = [sigma for x, y, sigma in array]
    
    plt.errorbar(
        xVals,
        yVals,
        yerr = sigmas,
        fmt = 'o',
        color = 'blue',
        ecolor = 'black' ,
        capsize = 3
    )
    plt.xlabel("X Axis")
    plt.ylabel("Y Axis")
    plt.title("Journal 2 Session 2")
    plt.savefig("J2S2 Graph.pdf")
    plt.show()
```

```python
if __name__ == "__main__":
    main()
```
